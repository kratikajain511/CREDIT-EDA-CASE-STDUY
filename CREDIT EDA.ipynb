```python
#Lets Check Code Gender Column first:
df1.CODE_GENDER.value_counts()
```

```python
#Replace the XNA with mode of the CODE_GENDER column:
(df1.CODE_GENDER.mode()[0])
```

```python
df1['CODE_GENDER'] = df1.CODE_GENDER.apply(lambda x : 'F' if x=="XNA" else x)

#Checking if the value is properly replaced:
df1.CODE_GENDER.value_counts()
```

```python
#Converting these values to (+)ve values:
df1["DAYS_BIRTH"] = df1["DAYS_BIRTH"].abs()
df1["DAYS_EMPLOYED"] = df1["DAYS_EMPLOYED"].abs()
df1["DAYS_REGISTRATION"] = df1["DAYS_REGISTRATION"].abs()
df1["DAYS_ID_PUBLISH"] = df1["DAYS_ID_PUBLISH"].abs()
df1["DAYS_LAST_PHONE_CHANGE"] = df1["DAYS_LAST_PHONE_CHANGE"].abs()

#Checking if the values are properly converted:
df1[["DAYS_BIRTH", "DAYS_EMPLOYED", "DAYS_REGISTRATION", "DAYS_ID_PUBLISH","DAYS_LAST_PHONE_CHANGE"]]
```

# SEGMENTATION OF COLUMNS

```python
print(df1.info(verbose = True, show_counts = True))
```

```python
for i in cat_cols:
    plt.figure(figsize=(10,5))
    sns.countplot(df1[i],data=target0)
    plt.title(i)
    plt.xticks(rotation=90)
    plt.show()
```

```python
for i in num_cols:
    plt.figure(figsize=(10,5))
    sns.histplot(df1[i], kde=True)
    plt.title(i)
    plt.xticks(rotation=90)
    plt.show()
```

# Bivariate Analysis of application data

```python
for i in cat_cols:
        for j in num_cols:
             sns.barplot(df1[i],df1[j]
        plt.xticks(rotation=90)
        plt.title("Barplot "+i+" Vs "+j)
        plt.show()
```
